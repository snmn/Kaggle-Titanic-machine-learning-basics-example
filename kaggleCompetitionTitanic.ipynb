```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/processed_titanic.csv')
data
```

```python
data['Age']
```

```python
X = data['Age']
I= data['PassengerId']
Y = data['Survived']
avg = X.mean()
X_a = np.array(X).reshape(-1,1)
Y_b = np.array(Y).reshape(-1,1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_a,Y_b)
```

```python

test_data = pd.read_csv('../data/titanicdata/test.csv')
new_test = test_data["Age"].fillna(avg)
new_Id = test_data["PassengerId"]
new_test
```

```python
new_Id
```

```python
test_data_array = np.array(new_test).reshape(-1,1)
new_model = model.predict(test_data_array)
new_model
```

```python
model.score(X_a,Y_b)
```
